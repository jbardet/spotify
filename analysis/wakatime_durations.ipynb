```python
from datetime import date, datetime as dt, timedelta as td
import numpy as np
import pandas as pd
import datetime

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
```

```python
durations = pd.read_csv('../data/wakatime_durations.csv')
# if exists remove old index columns
durations.drop(['Unnamed: 0'], axis=1, inplace=True)
durations
```

```python
# make a 1d bar with time and duration of each event for spotify
# first separate by day
spotify = durations[durations['project']=="spotify"]
spotify['date'] = pd.to_datetime(spotify['time']).dt.date
# spotify['date'] = spotify['date'].dt.date
# now get the time of the day
spotify['time'] = pd.to_datetime(spotify['time']).dt.time
spotify
```

```python
intervals = []
for i, row in spotify[spotify['date'] == datetime.datetime.strptime("2023-10-24", '%Y-%m-%d').date()].iterrows():
    timing_h = row['time'].hour+row['time'].minute/60+row['time'].second/3600
    intervals.append([timing_h, timing_h+row['duration']/3600])
```
